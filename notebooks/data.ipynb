```python
import time
import pandas as pd
import yfinance as yf
from pathlib import Path

tickers = ["SPY", "QQQ", "IWM", "IEF", "TLT"]

start_date = "2010-01-01"
end_date = "2024-12-31"

data_path = Path("../data")
data_path.mkdir(exist_ok=True)

prices = pd.DataFrame()

for ticker in tickers:
    print(f"Downloading {ticker}...")

    df = yf.download(
        ticker,
        start=start_date,
        end=end_date,
        auto_adjust=False,
        progress=False,
        threads=False
    )

    print(ticker, df.shape)

    if df.empty:
        print(f"{ticker} failed")
        continue

    df.to_csv(data_path / f"{ticker}.csv")

    prices[ticker] = df["Adj Close"] if "Adj Close" in df.columns else df["Close"]

    print(f"Saved {ticker}.csv")
    time.sleep(10)

prices = prices.sort_index().dropna()
prices.to_csv(data_path / "etf_prices_clean.csv")

print("Saved clean file:", data_path / "etf_prices_clean.csv")
print(prices.index.min(), "to", prices.index.max())
print(prices.shape)

prices.head()
```
